# Model & Tokenizer
NLP에서는 기본적으로 model과 tokenizer를 사용합니다.  
transformers 라이브러리는 다음의 순서로 진행합니다.  

1. 모델명 정하기  
2. 모델명에 맞는 model과 tokenizer 불러오기  
(model뿐 아니라 tokenizer도 pretrained 방식으로 이용합니다.)  

transformers의 AutoModel과 AutoTokenizer를 이용합니다.

```python
import torch
from transformers import AutoModel, AutoTokenizer

MODEL_NAME = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModel.from_pretrained(MODEL_NAME, num_labels=3) # labels = 0, 1, 2

# ImportError: IProgress not found. Please update jupyter and ipywidgets
# AutoModel... uses tqdm, needs ipywidgets / after installation, kernel needs to be restarted.

device = torch.device("cuda")
model.to(device)
model.device
```

```python
# 모델의 구조를 볼 수 있습니다.
a = model.parameters

# 모델의 config을 볼 수 있습니다.
# 또한 config을 변결할 수 있습니다.
b = model.config
#model.config.vocab_size = 119547

# 토크나이저 구성을 볼 수 있습니다.
c = tokenizer

b.num_labels
```

```python
def get_cls_token(sent_A):
    """
    BERT 모델에서 분류 결과는 [CLS] 토큰에 담깁니다.
    이 함수는 출력의 마지막 상태의 [CLS] 토큰을 반환합니다. (logits)
    
    logits의 shape은 (1, 768)입니다.
    768은 hidden size입니다. (BertConfig["hidden_size"] = 768)
    """
    tokenized_sent = tokenizer(
            sent_A,
            return_tensors="pt",
            truncation=True,
            add_special_tokens=True,
            max_length=128
    )
    tokenized_sent.to(device)

    model.eval()
    with torch.no_grad():
        #outputs = model(**tokenized_sent)
        outputs = model(
            input_ids=tokenized_sent['input_ids'],
            attention_mask=tokenized_sent['attention_mask'],
            token_type_ids=tokenized_sent['token_type_ids']
        )
    logits = outputs.last_hidden_state[:,0,:].detach().cpu().numpy()
    return logits
```

```python
# 우선 질문과 답변 문장들을 준비합니다.
chatbot_Question = ['기차 타고 여행 가고 싶어','꿈이 이루어질까?','내년에는 더 행복해질려고 이렇게 힘든가봅니다', '간만에 휴식 중', '오늘도 힘차게!'] # 질문
chatbot_Answer = ['꿈꾸던 여행이네요.','현실을 꿈처럼 만들어봐요.','더 행복해질 거예요.', '휴식도 필요하죠', '아자아자 화이팅!!'] # 답변
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 각 질문의 [CLS] 결과를 계산합니다. (dataset_cls_hidden)
# shape은 (5, 768)이며, 5는 질문의 개수입니다.
dataset_cls_hidden = []
for q in chatbot_Question:
    q_cls = get_cls_token(q)
    dataset_cls_hidden.append(q_cls)
dataset_cls_hidden = np.array(dataset_cls_hidden).squeeze(axis=1)
#print(dataset_cls_hidden)   # 데이터셋의 질문에 대한 [CLS] 토큰 벡터
print(dataset_cls_hidden.shape)
```

```python
dataset_cls_hidden
```

```python
query = '아 여행가고 싶다~'
query_cls_hidden = get_cls_token(query)

cos_sim = cosine_similarity(query_cls_hidden, dataset_cls_hidden)
print(cos_sim)

top_question = np.argmax(cos_sim)

print('나의 질문: ', query)
print('저장된 답변: ', chatbot_Answer[top_question])
```

```python
from tqdm import tqdm

new_dataset_cls_hidden = []
for q in tqdm(df["Q"]):
    q_cls = get_cls_token(q)
    new_dataset_cls_hidden.append(q_cls)
new_dataset_cls_hidden = np.array(new_dataset_cls_hidden).squeeze(axis=1)
```

```python
# 질문 선택
# 참고용
idx = 100
print(df["Q"][idx])
print(df["A"][idx])

query = "부자됐어"
```

```python
# CLS 토큰 및 코사인 유사도
query_cls_hidden = get_cls_token(query) # CLS 토큰
cos_sim = cosine_similarity(query_cls_hidden, new_dataset_cls_hidden) # 코사인 유사도
top_question = np.argmax(cos_sim)
print(top_question) # df["Q"]에서 query와 가장 비슷한 질문의 인덱스

print('나의 질문: ', query)
print('저장된 답변: ', df["A"][top_question])
```

```python
def get_chatting_input():
    query = input("you: ")
    if query == "q":
        return 0
    else:
        query_cls_hidden = get_cls_token(query) # CLS 토큰
        cos_sim = cosine_similarity(query_cls_hidden, new_dataset_cls_hidden) # 코사인 유사도
        top_question = np.argmax(cos_sim)

        print("bot: ", df["A"][top_question])
        return 1
```
